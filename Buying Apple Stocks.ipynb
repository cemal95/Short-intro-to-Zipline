## Buying Apple stocks quick example

```python
import zipline
from zipline.api import order_target_percent, symbol, record
import pytz
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
def initialize(context):
    context.stock = symbol('IWV')
    
def handle_data(context, data):
    apple_stock = context.stock
    order_target_percent(apple_stock, 1.0)

start_time = datetime(2010, 1, 1, tzinfo=pytz.UTC)
end_time = datetime(2016, 1, 1, tzinfo=pytz.UTC)

perf_Algo = zipline.run_algorithm(
    start=start_time,
    end=end_time,
    initialize=initialize,
    capital_base=100000,
    handle_data=handle_data,
    data_frequency='daily',
    bundle="backtest-bundle")
```

```python
port_val = perf_Algo.portfolio_value
plt.plot(port_val)
```
