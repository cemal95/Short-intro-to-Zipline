## Part 2 : Getting jupyter notebook to make things easier (also for developing)

Even if you already have Jupyter Notebook installed, in order to use Zipline in your notebooks, you need to use your virtual environment as your kernel. Open your environment through conda activate. Then use: **conda install nb_conda**, and then enter **jupyter notebook** in your command line. Doing so, it will open your browser with a main menu.
